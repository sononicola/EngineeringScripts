```python
import pandas as pd
```

```python
df = pd.read_pickle("test5.pkl")
df
```

```python
df.sort_values("location",inplace=True)
```

```python
for line in df["line_no"].unique():
    df_line = df.loc[
        (df["line_support_force_label"] == "max") & (df["line_no"] == line)
    ].copy()
    df_line.sort_values(by="location", inplace=True,ascending=True)
```

```python
df_line.plot(x="location", y="line_support_force_p_x")
```
